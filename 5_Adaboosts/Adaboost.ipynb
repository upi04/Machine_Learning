# Prepare Data

Load Data

1. split input-output
2. split train & test
3. handling problems in the training data
    - missing values
        - imputation
            -> numeric --> imput mean or median or mode
            -> categorical --> UNKNOWN or EMPTY, mode, clustering
        - drops
    - encode categorical variables --> change categorical to numeric
        - One hot encoding
        - encoding labels
            -> country name --> OHE
            -> rating -> label
    - scaling --> standard scaler.
4. Model fitting --> experimentation
    - baseline model
        - regression model --> average of output
        - classification model --> proportion of the largest class.
        
    - reference to help us choose the best model
        - performance metrics --> when we do cross validation.
        
    - knn -> number of neighbors, distance
    - tree -> max depth, criterion, min sample on leaf
    - linear model -> none
    - ridge & lasso -> alpha -> penalty regularization
    
5. prediction on test data -> sanity check performance on test data from the best model.

```python
def scaler_transform(X, scaler = None):
    """
    Fungsi untuk melakukan standardisasi
    :param X: <pandas DataFrame> sampel data
    :param scaler: <sklearn object> scaler, default None
    :return X_scaled: <pandas Dataframe> sampel data OHE
    :param scaler: <sklearn object> scaler, default None
    """
    if scaler != None:
        pass
    else:
        # Buat & fit encoder
        scaler = StandardScaler()
        scaler.fit(X)

    # Tranform data
    X_scaled = scaler.transform(X)
    X_scaled = pd.DataFrame(X_scaled,
                            columns = X.columns,
                            index = X.index)

    return X_scaled, scaler
```

```python
def prepare_test(X, scaler):
    # Scaled data
    X_clean, _ = scaler_transform(X = X,
                                  scaler = scaler)

    return X_clean
```

```python
def prepare_test(X, scaler):
    # Scaled data
    X_clean, _ = scaler_transform(X = X,
                                  scaler = scaler)

    return X_clean
```

# Bagging Method

1. Membuat $B$-sampel bootstrap
2. Membuat $B$-model decision tree untuk masing-masing sampel
3. Setiap model melakukan prediksi.
4. Melakukan agregasi dari setiap hasil prediksi.
    - Regresi menggunakan rata-rata
    - Klasifikasi menggunakan majority vote

```python
# Fungsi Bagging
def bagging(X_train, y_train, B):
    """
    Fungsi untuk membuat model bagging
    :param X_train: <pandas Dataframe> sampel input
    :param y_train: <pandas Dataframe> sampel output
    :param B: <int> jumlah bootstrap sample
    :return trees: <list> kumpulan tree
    """
    N = len(X_train)
    trees = []

    for b in range(B):
        # Buat bootstrap
        sample = np.random.choice(X_train.index,
                                  size = N,
                                  replace = True)

        X_train_b = X_train.loc[sample]
        y_train_b = y_train.loc[sample]

        # Buat model
        tree = DecisionTreeRegressor()
        tree.fit(X_train_b, y_train_b)

        # Append tree
        trees.append(tree)

    return trees
```

```python
bagging_mdl = bagging(X_train = X_train_clean,
                      y_train = y_train,
                      B = 5)

bagging_mdl
```

```python
# Fungsi predict
def predict_bagging(estimator, X):
    """
    Fungsi untuk melakukan prediksi
    :param estimator: <list> list model
    :param X: <pandas Dataframe> sampel input
    :param y_pred: <pandas Dataframe> prediksi sampel output
    """
    B = len(estimator)
    N = X.shape[0]
    y_pred_list = np.zeros((B, N))

    for i, tree in enumerate(estimator):
        # Predict
        y_pred_list[i] = tree.predict(X)

    # Average
    y_pred = np.mean(y_pred_list, axis = 0)

    return y_pred
```

```python
# Predict
y_pred_train = predict_bagging(estimator = bagging_mdl,
                               X = X_train_clean)
y_pred_test = predict_bagging(estimator = bagging_mdl,
                              X = X_test_clean)
```

```python
# Calculate MAE
mae_train = mean_absolute_error(y_train, y_pred_train)
mae_test = mean_absolute_error(y_test, y_pred_test)

mae_train, mae_test
```

```python
# Lakukan perulangan
B = [10, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
mae_train_list = []
mae_test_list = []

for b in B:
    # Buat bagging model
    bagging_mdl = bagging(X_train = X_train_clean,
                          y_train = y_train,
                          B = b)

    # Predict
    y_pred_train = predict_bagging(estimator = bagging_mdl,
                                   X = X_train_clean)
    y_pred_test = predict_bagging(estimator = bagging_mdl,
                                  X = X_test_clean)

    # Calculate mae
    mae_train = mean_absolute_error(y_train, y_pred_train)
    mae_test = mean_absolute_error(y_test, y_pred_test)

    # Append MSE
    mae_train_list.append(mae_train)
    mae_test_list.append(mae_test)
```

```python
# Plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 7))

ax.plot(B, mae_train_list, "b", marker = "o", label = "Train")
ax.plot(B, mae_test_list, "r", marker = "x", label = "Test")

ax.set_xlabel("B")
ax.set_ylabel("MAE")
plt.legend()
plt.show()
```

<font color='red'>
Notes! - The image above only shows that the greater the B value, the model will not overfit.

The selection of the best model is <b>not</b> done by looking at test performance, <b>but</b> by looking at CV performance.
</font>

# Random Forest Method

1. Create $B$-bootstrap samples
2. Determine $p$ features used to create the Decision Tree model
3. Create a $B$-model decision tree for each sample
4. Each model makes predictions.
5. Aggregate each prediction result.
    - Regression uses the average
    - Classification uses majority vote

```python
from sklearn.ensemble import RandomForestRegressor
```
